```python
import numpy as np 
chess = np.zeros ((800, 800))
```

```python
print(chess)
```

```python
for i in range (0 , 800, 100):
  for j in range (0 , 800 , 100):
    if i%2==0 and j%2==0:
      chess [i:i+100 , j:j+100]= 255
    elif  i%2!=0 and j%2!=0:
      chess [i:i+100 , j:j+100]= 255
    elif i%2==0 and j%2!=0 :
      chess [i:i+100 , j:j+100]= 0
    elif i%2!=0 and j%2==0 :
       chess [i:i+100 , j:j+100]= 0
```

```python
from google.colab.patches import cv2_imshow
```

```python
print(chess)
```

```python
cv2_imshow(chess)
```
