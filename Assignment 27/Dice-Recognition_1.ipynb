```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Input/0.jpg', cv2.IMREAD_GRAYSCALE)
img_origin = img.copy()
```

```python
_, img = cv2.threshold(img, 70, 255, cv2.THRESH_BINARY)
img = 255 - img
plt.imshow(img, cmap = "gray")
```

```python
contours , _ = cv2.findContours(img, cv2.RETR_LIST, cv2.CHAIN_APPROX_NONE)
score = 0
for c in contours:
    if len(c) > 100:
        x, y, w, h = cv2.boundingRect(c)
        cv2.rectangle(img_origin, (x,y) , (x+w,y+h), (0, 255, 0), 4)
        score += 1

cv2.putText(img_origin, f'score : {score}', (30, 70), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0),thickness=2)
plt.imshow(img_origin, cmap = "gray")
cv2.imwrite('Output/2.jpg', img_origin)
```
