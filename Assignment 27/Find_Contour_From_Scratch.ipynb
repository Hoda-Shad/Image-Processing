```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
imgOriginal = cv2.imread('Input/hat.jpg')
imgGray = cv.cvtColor(imgOriginal, cv.COLOR_BGR2GRAY)
print(imgGray.shape)
plt.imshow(imgGray, cmap="gray")
```

```python
imgGray=255-imgGray
plt.imshow(imgGray,cmap='gray')
```

```python
_,imgGray=cv2.threshold(imgGray,95,255,cv2.THRESH_BINARY)
plt.imshow(imgGray,cmap='gray')
```

```python
height,width = imgGray.shape[0:2]

contour_index=[]
for i in range(height):
    for j in range(width):
        # print(imgGray[i,j])
        if (imgGray[i,j]!= 0):
            if((imgGray[i,j+1]==0) or (imgGray[i,j-1]==0) or (imgGray[i+1,j]==0) or (imgGray[i-1,j]==0)):
                contour_index.append([(j,i)])
points=np.array(contour_index)
#print(points)
               
```

```python
cv2.drawContours(imgGray,points,-1,(60,25,40),10)
plt.imshow(imgGray,cmap='gray')
cv2.imwrite('Output/hat.jpg', imgGray)
```
