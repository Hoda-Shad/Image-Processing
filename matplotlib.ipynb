```python
##Scatter
```

```python
x = np.array ([10,20,30,40,18,26,15])
y = np.array ([80,90,10,12,13,36,26])
plt.scatter(x,y, color="purple")

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = "red")

plt.show()
```

```python
x = np.array ([10,20,30,40,18,26,15])
y = np.array ([80,90,10,12,13,36,26])
colors = np.array(['pink','red','green','blue','purple','black','yellow'])
plt.scatter(x,y, color = colors)
plt.show()
```

```python
x = np.array ([10,20,30,40,18,26,15])
y = np.array ([80,90,10,12,13,36,26])
colors = ([0,10,20,30,40,50,60])
plt.scatter(x,y , c = colors , cmap = 'Spectral')
```

```python
import random
x = np.random.randint((100), size = (100))
y = np.random.randint((100), size = (100))
colors = np.random.randint((100), size = (100))
plt.scatter(x,y, c= colors , cmap = 'Accent')
plt.colorbar()
```
