```python
import cv2
Ronaldo = cv2.imread('Ronaldo2.jpg', 0)
Hoda = cv2.imread('Hoda.jpg', 0)
```

```python
Hoda.shape
```

```python

resized = cv2.resize(Hoda, (199,253))
```

```python
resized.shape
```

```python
result1 = resized//2 + Ronaldo//8
```

```python
result2 = resized//2 + Ronaldo//4
```

```python
result3 = resized//2 + Ronaldo//2
```

```python
result4 = resized//8 + Ronaldo//2
```

```python
result5 = resized//4 + Ronaldo//2
```

```python

import cv2
import numpy as np
from skimage.util import random_noise
from google.colab.patches import cv2_imshow
img = cv2.imread('Ronaldo.jpg',0)
print(img)
noise_img = random_noise(img, mode='s&p',amount=0.1)
noise_img = np.array(255*noise_img, dtype = 'uint8')
cv2_imshow(noise_img)
cv2.imwrite('noisy_image.jpg', noise_img)
```

```python
import cv2
```

```python
img = cv2 .imread ('Mona_Lisa.jpg', 0)
inverted = 255 - img
```

```python
cv2_imshow(inverted)
```

```python
blurred = cv2.GaussianBlur(inverted , (21,21) , 0)
```

```python
inverted_blurred = 255- blurred
```

```python
cv2_imshow(inverted_blurred)
```

```python
sketch = img / inverted_blurred
```

```python
sketch = sketch * 255
```

```python
cv2_imshow(sketch)
```

```python
cv2.imwrite('pen_pic.jpg', sketch)
```
